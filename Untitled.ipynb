```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt

#define system
omega = 1 # rad/s

#define plot
N_per = 5
T = 2*np.pi/omega*N_per
N = 1000
t = np.linspace(0,T,N)
x = np.zeros((1,N))

A1 = 1
phi1 = 0
A2 = 2
phi2 = np.pi*3/4

#
x1 = A1*np.cos(omega*t + phi1)
x2 = A2*np.cos(omega*t + phi2)
x3 = x1 + x2
plt.plot(t, x1, color='blue', linestyle='dashed')
plt.plot(t, x2, color='red', linestyle='dashed')
plt.plot(t,x3, color='black', linestyle='dashed')
plt.xlabel('t')
plt.grid()
```
